```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import networkx as nx
from sklearn.utils import shuffle
from sklearn import preprocessing
```

```python
!wget https://raw.githubusercontent.com/RosalRicardo/tweet-midia/master/dataset_followers_midia_2.csv
```

```python
df_followers = pd.read_csv('dataset_followers_midia_2.csv')
df_followers = df_followers[['screen_name','ref']]
df_followers = shuffle(df_followers)
df_followers.head(10)
```

```python
le = preprocessing.LabelEncoder()
le.fit(df_followers['ref'])
le.transform(df_followers['ref'])
df_followers['ref_id'] = le.transform(df_followers['ref'])
#df_followers.ref_id.unique()
df_teste = df_followers.head(2)
```

```python

G = nx.Graph()
G = nx.from_pandas_edgelist(df_teste, 'ref', 'screen_name')

#configuração do plot
plt.rcParams["figure.figsize"] = [20, 15]
plt.rcParams["figure.autolayout"] = True
nx.draw(G,with_labels=False, node_size=5, node_color='g', alpha=1, linewidths=5)
plt.show()
```

```python
G = nx.Graph()
G = nx.from_pandas_edgelist(df_teste, 'ref', 'screen_name')

#configuração do plot
plt.rcParams["figure.figsize"] = [20, 15]
plt.rcParams["figure.autolayout"] = True
nx.draw_circular(G,with_labels=True, node_size=5, node_color='g', alpha=1, linewidths=5)
plt.show()
```
